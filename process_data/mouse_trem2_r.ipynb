## GSE200275

```r
object <- qread("/lustre/groups/ml01/projects/2024_microglia_zihe.zheng/mouse_trem2/GSE200275_integrated.qs")
```

```r
object
```

```r
DefaultAssay(object) <- 'RNA'
```

```r
object
```

```r
34737+32532+34774
```

```r
counts1 <- LayerData(object[["RNA"]], layer = "counts.1")
```

```r
counts2 <- LayerData(object[["RNA"]], layer = "counts.2")
```

```r
counts3 <- LayerData(object[["RNA"]], layer = "counts.3")
```

```r
counts = cbind(counts1, counts2, counts3)
```

```r
obj_v4 <- CreateSeuratObject(counts = counts)

# Transfer metadata
obj_v4@meta.data <- object@meta.data
```

```r
obj_v4[["RNA"]]@counts
```

```r
SaveH5Seurat(obj_v4, filename = "/lustre/groups/ml01/projects/2024_microglia_zihe.zheng/mouse_trem2/GSE200275_integrated.h5Seurat", assays = c("RNA", "integrated"), overwrite = TRUE)
```

```r
Convert("/lustre/groups/ml01/projects/2024_microglia_zihe.zheng/mouse_trem2/GSE200275_integrated.h5Seurat", dest = "h5ad", overwrite = TRUE)
```

## GSE209912

```r
object <- qread("/lustre/groups/ml01/projects/2024_microglia_zihe.zheng/mouse_trem2/GSE209912_integrated.qs")
```

```r
object
```

```r
# LayerData(object[["RNA"]], layer = "scale.data")
```

```r
# Extract the layers from the RNA assay
counts <- LayerData(object[["RNA"]], layer = "counts")
data   <- LayerData(object[["RNA"]], layer = "data")
scale  <- LayerData(object[["RNA"]], layer = "scale.data")
```

```r
obj_v4 <- CreateSeuratObject(counts = counts)

# Transfer metadata
obj_v4@meta.data <- object@meta.data

# Transfer dimensional reductions if needed
obj_v4@reductions <- object@reductions
```

```r
obj_v4@version
```

```r
obj_v4[["RNA"]]@counts
```

```r
SaveH5Seurat(obj_v4, filename = "/lustre/groups/ml01/projects/2024_microglia_zihe.zheng/mouse_trem2/GSE209912_integrated.h5Seurat", assays = c("RNA"), overwrite = TRUE)
```
