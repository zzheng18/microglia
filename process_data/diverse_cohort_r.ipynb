```r
library(Seurat)
library(SeuratDisk)
```

```r
cell_types <- c('Astro', 'Endo', 'Epend', 'Gaba', 'Glut', 'Micro', 'Olig', 'Opc', 'OtherImmune')

for (cell_type in cell_types) {
    rds_path <- paste0('/lustre/groups/ml01/projects/2024_microglia_zihe.zheng/diverse_cohort/counts_', cell_type, '_filtered.rds')
    h5seurat_path <- paste0('/lustre/groups/ml01/projects/2024_microglia_zihe.zheng/diverse_cohort/counts_', cell_type, '_filtered.h5Seurat')
    h5ad_path <- paste0('/lustre/groups/ml01/projects/2024_microglia_zihe.zheng/diverse_cohort/counts_', cell_type, '_filtered.h5ad')

    seurat_obj <- readRDS(rds_path)
    seurat_obj <- CreateSeuratObject(seurat_obj, project = cell_type, assay = "RNA")

    SaveH5Seurat(seurat_obj, filename = h5seurat_path)

    Convert(h5seurat_path, dest = "h5ad")
}
```
