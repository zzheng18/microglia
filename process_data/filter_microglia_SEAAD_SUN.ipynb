## Sun ROSMAP (filter for only microglia states)

```python
rna_sun = sc.read_h5ad('/home/icb/zihe.zheng/projects/microglia/data/SUN/MIT_ROSMAP_RNA.h5ad')
```

```python
meta = pd.read_csv("/home/icb/zihe.zheng/projects/microglia/data/meta.tsv", sep="\t")
```

```python
state_dict = meta.set_index('cellName')['State'].to_dict()
rna_sun.obs['State'] = rna_sun.obs.index.map(state_dict)
```

```python
rna_sun_filtered = rna_sun[~rna_sun.obs['State'].isna()].copy()
```

```python
rna_sun_filtered.var = rna_sun_filtered.var.set_index('_index', drop = True)
rna_sun_filtered.var.index.name = None
```

```python
adata_cleaned = sc.AnnData(X=rna_sun_filtered.X, obs=rna_sun_filtered.obs.copy(), var=rna_sun_filtered.var.copy())
```

```python
adata_cleaned.write_h5ad('/home/icb/zihe.zheng/projects/microglia/data/SUN/microglia_MIT_ROSMAP_RNA.h5ad')
```
