```python
import scanpy as sc
import pandas as pd
import os
import glob
```

```python
data_dir = "/lustre/groups/ml01/projects/2024_microglia_zihe.zheng/mouse_glp1ra"
```

```python
expr_path = os.path.join(data_dir, "expression_full.csv")
meta_path = os.path.join(data_dir, "meta.tsv")
expr = pd.read_csv(expr_path, index_col=0)
meta = pd.read_csv(meta_path, sep='\t', index_col=0)
```

```python
expr
```

```python
meta.drop(index='TYPE', inplace=True)
```

```python
if set(meta.index) == set(expr.columns):
    expr = expr.T
elif set(meta.index) != set(expr.index):
    raise ValueError("Metadata and expression IDs don't match.")
```

```python
adata = sc.AnnData(X=expr)
adata.obs = meta.loc[expr.index]
```

```python
adata.obs
```

```python
adata
```

```python
adata.obs['nCount_RNA'] = pd.to_numeric(adata.obs['nCount_RNA'])
adata.obs['nFeature_RNA'] = pd.to_numeric(adata.obs['nFeature_RNA'])
adata.obs['percent_mt'] = pd.to_numeric(adata.obs['percent_mt'])
adata.obs['seurat_clusters'] = pd.to_numeric(adata.obs['seurat_clusters'])
```

```python
adata.write(os.path.join(data_dir, "adata.h5ad"))
```
